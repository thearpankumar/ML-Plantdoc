```python
import numpy as np 
import pandas as pd
```

```python
df=pd.read_csv("dataset.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop([" OPEN_PRICE"," HIGH_PRICE"," LOW_PRICE"," LAST_PRICE"," AVG_PRICE"," TTL_TRD_QNTY"," TURNOVER_LACS"," NO_OF_TRADES"," DELIV_QTY"," DELIV_PER"], axis=1)
df["change"]=(((df[" CLOSE_PRICE"]-df[" PREV_CLOSE"])/df[" PREV_CLOSE"])*100)
```

```python
df[' DATE1'] = pd.to_datetime(df[' DATE1'], dayfirst=True)
df['month'] = pd.DatetimeIndex(df[' DATE1']).month
df['year'] = pd.DatetimeIndex(df[' DATE1']).year
df=df.sort_values(by=["SYMBOL","month","year"])
```

```python
#pip install pandasql
```

```python
from pandasql import sqldf
q="select SYMBOL,month,year,avg(change) as change from df group by symbol,year,month order by SYMBOL"
newdf=sqldf(q, globals())
newdf
```

```python
newdf=newdf.sort_values(by=["year","month","change"])
newdf
```

```python
#Top performing Stock
q1="select symbol,month,year,max(change) from newdf group by year,month"
sqldf(q1, globals())
```

```python
#Top Non Performing Stock
q2="select symbol,month,year,min(change) from newdf group by year,month"
sqldf(q2, globals())
```
