#### Importing Dataset and grouping data that were bought by same customer on same date..

```
dataset=pd.read_csv('Groceries_dataset.csv')
grouped=dataset.groupby(by=['Member_number','Date'])
group=grouped.indices
```

```
print(group)
```

### Extracting keys of group dictionary 

```
ky=[]
for i,j in group.items():
  ky.append(i)
```

### Making List of grouped itemDescription that bought by same member on same day

```
transactions=[]
transactions.append([grouped.get_group(i)['itemDescription'].tolist() for i in ky])
```

```
print(transactions)
```

### Converting 3D list into 2D list

```
transaction=[]
for i in transactions:
  for j in i:
    transaction.append(j)
```

```
print(transaction)
```

## Training apriori model on dataset

```
from apyori import apriori
rules=apriori(transactions=transaction,min_support=0.00030, min_confidance=0.01,
               min_lift=3, min_length=2, target="rules")
```

```
print(rules)
```

### Displaying first result directly comming from apriori function

```
result=list(rules)
print(result)
```

## Putting result well organised in pandas DataFrame

```
def inspect(result):
  lhs=[tuple(res[2][0][0])[0] for res in result]
  rhs=[tuple(res[2][0][1])[0] for res in result]
  support=[res[1] for res in result]
  confidence=[res[2][0][2] for res in result]
  lift=[res[2][0][3] for res in result]
  return(list(zip(lhs,rhs,support,confidence,lift)))
```

```
resultinDataFrame=pd.DataFrame(inspect(result),columns=['Left Hand Side','Right Hand Side','Support','Confidence','Lift'])
```

## Displaying the result according to sorted lift

```
resultinDataFrame.nlargest(n=15,columns='Lift')
```
